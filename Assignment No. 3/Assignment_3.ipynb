#####  Problem Statement 1. Create a CSV file containing product names and prices. Read the data using Pandas.
##### Use NumPy to calculate the average price. Write a function to identify products priced above the average
##### and display them.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("products.csv")

average = np.mean(data["Price"])

print("Average Price:", average)

def above_average():
    print("Products above average price:")
    print(data[data["Price"] > average])

above_average()
```

#### Problem Statement 2. Read daily temperature data from a CSV file using Pandas. Use NumPy to calculate 
#### highest, lowest, and average temperatures. Write functions to classify each day as: Cold (&lt;20°C),
#### Moderate (20–30°C), Hot (&gt;30°C).Display the category for each day.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("temperature.csv")

highest = np.max(data["Temperature"])
lowest = np.min(data["Temperature"])
average = np.mean(data["Temperature"])

print("Highest Temperature :", highest)
print("Lowest Temperature :", lowest)
print("Average Temperature :", average)

def category(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"

print("Temperature Category")

for i in range(len(data)):
    print(data["Day"][i], ":", category(data["Temperature"][i]))
```
