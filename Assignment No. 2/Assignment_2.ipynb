 ##### Question 1. Create a program that reads student marks from a file (filename.txt). Display
 ##### the names of students who have passed (marks &gt; 45) and those who have failed. Count the 
 ##### total number of passed and failed students.

```python
file = open("filename.txt", "r")

pass_count = 0
fail_count = 0

print("Passed Students:")

for i in file:
    data = i.split()
    name = data[0]
    marks = int(data[1])

    if marks > 45:
        print(name)
        pass_count += 1
    else:
        fail_count += 1

file.close()

print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

#### Question 2. A file (salary.txt) contains employee names and salaries. Read the file and
#### display only those employees whose salary is greater than ₹50,000.
#### Also count how many employees satisfy this condition.

```python
file = open("salary.txt", "r")

count = 0

print("Employees with salary greater than 50000:")

for i in file:
    data = i.split()

    name = data[0]
    salary = int(data[1])

    if salary > 50000:
        print(name)
        count = count + 1

file.close()

print("Total Employees:", count)
```

#### Question 3. A file (attendance.txt) contains the attendance status of students(Present/Absent).
#### Read the file and display only the names of absent students. 
#### Also calculate the total number of present and absent students.

```python
file = open("attendance.txt", "r")

present = 0
absent = 0

print("Absent Students:")

for i in file:
    data = i.split()

    name = data[0]
    status = data[1]

    if status == "Absent":
        print(name)
        absent = absent + 1
    else:
        present = present + 1

file.close()

print("Total Present Students:", present)
print("Total Absent Students:", absent)
```
