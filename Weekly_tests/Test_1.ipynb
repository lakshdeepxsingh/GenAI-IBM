### Basic Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### Given Dataset

```python
X = [1, 2, 3, 4, 5, None, 7, 8, 9, 10, 11, 12, None, 14, 15, 16, 17, 18, 19, 20]
```

```python
y_reg = [2, 4, 6, 8, 10, None, 14, 16, 18, 20, 22, 24, None, 28, 30, 32, 34, 36, 38, 40]
```

```python
y_cls = [0, 0, 0, 0, 0, None, 0, 1, 1, 1, 1, 1, None, 1, 1, 1, 1, 1, 1, 1]
```

### TASK 1 (PLOTTING SCATTER CHART)

```python
# TASK 1

plt.scatter(X,y_reg,color="b")
plt.xlabel("X")
plt.ylabel("Y_reg")
plt.title("Scatter Plot for X and y_reg")
plt.show()
```

### TASK 2 (LINEAR REGRESSION)

```python
# TASK 2       (LINEAR REGRESSION MODEL)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
```

```python
df1 = pd.DataFrame(X)
df2 = pd.DataFrame(y_reg)
```

```python
df1.isnull().sum()
```

```python
df3 = df1.dropna()
```

```python
df4 = df2.dropna()
```

```python
x = df3
y = df4
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=25)
```

```python
model =LinearRegression()
```

```python
m1 = model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error",mae)
```

```python
mse = mean_squared_error(y_test,y_pred)
print("Mean Squared Error :",mse)
```

```python
final_prediction1 = model.predict([[7]])
print("Final Prediction:",final_prediction1)
```

### TASK 3 (LOGISTIC REGRESSION)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,f1_score
```

```python
df5 = pd.DataFrame(X)
df6 = pd.DataFrame(y_cls)
```

```python
df7 = df5.dropna()
df8 = df6.dropna()
```

```python
x1 = df7
y1 = df8
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.2,train_size=0.8,random_state=25)
```

```python
model = LogisticRegression()
```

```python
m2 = model.fit(x1_train,y1_train)
```

```python
y1_pred = model.predict(x1_test)
```

```python
acc1 = accuracy_score(y1_test,y1_pred)
print("Accuracy:",acc1)
```

```python
f1 = f1_score(y1_test,y1_pred)
print("F1-Score: ",f1)
```

```python
fp = model.predict([[4]])
print("Final Prediction:",fp)
```
