```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv("/content/drive/MyDrive/dataset1.csv")
```

```python
df.head()
```

### Tokenization

```python
df["Tokens"] = df["Message"].apply(word_tokenize)
```

```python
df.sample(3)
```

### Stop Word Removal

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = []

for token in df["Tokens"]:
    temp = []

    for word in token:
        if word.lower() not in stop_words:
            temp.append(word)

    filtered_tokens.append(temp)

df["Filtered_Tokens"] = filtered_tokens

print(df["Filtered_Tokens"])
```

### Stemming

```python
ps = PorterStemmer()
stemmed_tokens = []

for token in df["Filtered_Tokens"]:

    temp = []

    for word in token:
        temp.append(ps.stem(word))

    stemmed_tokens.append(temp)

df["Stemmed_Tokens"] = stemmed_tokens

df['Stemmed_Tokens']
```
